```python
import pandas as pd
import numpy as np
from random import randint, choice
from faker import Faker
from datetime import datetime, timedelta
import itertools
```

```python

fake = Faker()

# Número de nomes únicos para cada coluna
num_names = [350, 800, 1300, 2500, 3800, 5200, 8500, 15000]

# Crie um conjunto para armazenar todos os nomes já usados
used_names = set()

# Gere listas de nomes únicos para cada coluna
names = []
for n in num_names:
    name_list = []
    for _ in range(n):
        name = fake.name()
        # Se o nome já foi usado, adicione um numerador a ele
        while name in used_names:
            name += str(len(name_list))
        name_list.append(name)
        used_names.add(name)
    names.append(name_list)

# Crie um dataframe vazio com 15 mil linhas
df = pd.DataFrame(index=range(max(num_names)))

# Preencha cada coluna do dataframe com seus respectivos nomes
for i, name_list in enumerate(names):
    # Repita a lista de nomes até que ela tenha o mesmo comprimento que o dataframe
    repeated_names = list(itertools.islice(itertools.cycle(name_list), len(df)))
    df[f'Nível_{i+1}'] = repeated_names

df
```

```python

# Definindo o número de linhas
num_linhas = 3200000

# Gerando uma lista de vendedores a partir do DataFrame anterior
vendedores = df['Nível_8'].tolist()

# Gerando datas de venda aleatórias de 2022 até hoje
start_date = datetime(2022, 1, 1)
end_date = datetime.today()
date_range = (end_date - start_date).days
datas_venda = [fake.date_between(start_date=start_date, end_date=end_date) for _ in range(num_linhas)]

# Gerando IDs de produtos, quantidades de itens e valores de itens aleatórios
ids_produto = [randint(1, 1000) for _ in range(num_linhas)]
qtde_itens = [randint(1, 100) for _ in range(num_linhas)]
valor_item = [round(np.random.uniform(1, 100), 2) for _ in range(num_linhas)]

# Criando o DataFrame de vendas
df_vendas = pd.DataFrame({
    'Data da Venda': datas_venda,
    'ID Produto': ids_produto,
    'Qtde Itens': qtde_itens,
    'Valor Item': valor_item,
    'Vendedor': np.random.choice(vendedores, num_linhas)
})

```
